```python
msft = yf.Ticker('MSFT') 
info_1y = msft.history(period = "1y")
info_1y['Date'] = info_1y.index
info_1y = info_1y.reset_index (drop = True)
```

```python
#Adding a column to calculate dif between Open and Close rates
info_1y['Dif_open_close'] = info_1y['Open'] - info_1y['Close']
info_1y.head()
```

```python
#Adding a column to calculate variation in that date
info_1y['Variation'] = info_1y['High'] - info_1y['Low']
info_1y.head()
```

```python
#create connection to redshift
conn = redshift_connector.connect(
    host='aaa',
    database='aaa',
    port=int('aaa'),
    user='aaa',
    password='aaa'
)
```

```python
cursor = conn.cursor()
```

```python
#def to execute query into redshift
def redshift_execute(conn,query):
    try:
        cur = conn.cursor()
        cur.execute(query)
        conn.commit()
        cur.close()
    except Exception as err:
        print (err)
```

```python
#Rollback to roll back the prev. query
redshift_execute(conn, "ROLLBACK;")
```

```python
# query to create table called entregable1 in schema asy_antonioyoo_coderhouse
query = f"""CREATE TABLE asy_antonioyoo_coderhouse.entregable1 (Open decimal(10,6), High decimal(10,6), Low decimal(10,6), Close decimal(10,6), volume Integer, Dividends decimal(10,6), Stock_splits decimal(10,6), date timestamp)"""
redshift_execute(conn,query)
```

```python
#Alter table to add two more columns created before
query = f"""
    ALTER TABLE asy_antonioyoo_coderhouse.entregable1
    ADD COLUMN Dif_open_close decimal(10,6)
"""
redshift_execute(conn,query)
```
